```python
import os,cv2
import numpy as np
from glob import glob
```

```python
train_dir = r"./det_bookspine_text/train/*.jpg"
train_path_l = glob(train_dir)
im_h_l, im_w_l = [], []
for i in train_path_l:
   # print(i)
   im = cv2.imread(i)
   h, w = im.shape[:2]
   im_h_l.append(h)
   im_w_l.append(w)

# 计算宽度和高度的集中范围（例如：取中间的80%数据）
width_sorted = np.sort(im_w_l)
height_sorted = np.sort(im_h_l)
width_10th, width_90th = width_sorted[int(0.1 * len(im_w_l))], width_sorted[int(0.9 * len(im_w_l))]
height_10th, height_90th = height_sorted[int(0.1 * len(im_h_l))], height_sorted[int(0.9 * len(im_h_l))]

print(f"宽度集中范围（10%-90%）：{width_10th} - {width_90th}")
print(f"高度集中范围（10%-90%）：{height_10th} - {height_90th}")

# # 绘制宽度和高度的直方图
# plot_histogram(widths, "图像宽度分布", "宽度 (像素)", "图像数量")
# plot_histogram(heights, "图像高度分布", "高度 (像素)", "图像数量")
```
